```python
import pandas as pd
import numpy as np
```

```python
# 1. تحميل بيانات Cleveland الخام من UCI
column_names = [
    'age', 'sex', 'cp', 'trestbps', 'chol',
    'fbs', 'restecg', 'thalach', 'exang',
    'oldpeak', 'slope', 'ca', 'thal', 'target'
]
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
url = "../datasets//processed.cleveland.data"

# قراءة الملف (احرص على وضع الملف processed.cleveland.data في نفس المجلد)
df = pd.read_csv(url, header=None, names=column_names,na_values='?')

# # 2. استبدال القيم المفقودة المشار إليها بـ '?' إلى NaN
# df.replace('?', np.nan, inplace=True)

# 3. نظرة سريعة على القيم الأولية
print(df.head())
print()
# 3. نظرة سريعة على القيم المفقودة
print(df.isnull().sum())
```

```python
df.info()
```

```python
# 3. تحويل الأعمدة التي تحتوي على '?'
cols_to_convert = ['ca', 'thal']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 4. حذف الصفوف التي تحتوي على قيم مفقودة
df.dropna(inplace=True)
```

```python
df.info()
```

```python
# 5. تحويل جميع الأعمدة لأنواعها الصحيحة
convert_types = {
    'age': float,
    'sex': int,
    'cp': int,
    'trestbps': float,
    'chol': float,
    'fbs': int,
    'restecg': int,
    'thalach': float,
    'exang': int,
    'oldpeak': float,
    'slope': int,
    'ca': int,
    'thal': int,
    'target': int
}
df = df.astype(convert_types)
```

```python
# 6. تحويل العمود 'condition' إلى 0 = لا مرض، 1 = يوجد مرض
# لأن القيم الأصلية: 0 = لا مرض، 1-4 = يوجد مرض قلب
df['target'] = df['target'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# 8. معاينة أول 5 صفوف
print("تم تنظيف البيانات بنجاح. أول 5 صفوف:")
print(df.head())
```

```python
# ترتيب القيم  من العمر الأصغر الى الأكبر
df = df.sort_values(by='age', ascending=False).reset_index(drop=True)
print(df.head())

# 7. حفظ الملف إلى تنسيق CSV النهائي
output_filename = '../datasets/heart_disease_cleaned.csv'
df.to_csv(output_filename, index=False)
```
